# Import the Packages

```python
import gradio
from groq import Groq
```

```python
client = Groq(
    api_key="Your_API_key",
)
```

#Define a function to give content and role

```python
def initialize_messages():
    return [{
        "role": "system",
        "content": (
            "You are an expert college admission assistant for students in India who have completed Class 12th. "
            "You provide accurate, up-to-date, and helpful guidance related to college admissions. You can answer queries about:\n"
            "- Available courses after Class 12th (engineering, medical, arts, commerce, etc.)\n"
            "- Colleges offering specific courses across India\n"
            "- Admission eligibility and prerequisites for each course\n"
            "- Rank cut-offs for various colleges and courses based on entrance exams (JEE, NEET, CUET, etc.)\n"
            "- NIRF rankings of colleges\n"
            "- Whether a college is government or private\n"
            "- College fee structures and scholarship options\n"
            "- Hostel facilities, campus amenities, and student life\n"
            "- Placement records and career scope for different courses\n"
            "- Special entrance exams required by individual colleges\n"
            "- Popular colleges with high placement rates and good reputations\n"
            "- Location and contact information of colleges\n"
            "- Admission process and important deadlines\n"
            "Answer clearly, factually, and in a student-friendly tone. Focus only on queries related to Indian colleges and admissions after Class 12th."
        )
    }]
```

#Assign it to a variable

```python
messages_prmt = initialize_messages()
```

```python
print(type(messages_prmt))
```

```python
[{},{}]
```

#Define a function to connect with LLM

```python
def customLLMBot(user_input, history):
    global messages_prmt

    messages_prmt.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        messages=messages_prmt,
        model="llama3-8b-8192",
    )
    print(response)
    LLM_reply = response.choices[0].message.content
    messages_prmt.append({"role": "assistant", "content": LLM_reply})

    return LLM_reply
```

#Create an object of chat interface class in gradio

```python
iface = gradio.ChatInterface(
    fn=customLLMBot,
    chatbot=gradio.Chatbot(height=300),
    textbox=gradio.Textbox(placeholder="Ask your question about Indian college admissions"),
    title="ğŸ“ College Admission Assistant",
    description="Get help with Indian colleges, cutoffs, courses, rankings, hostel facilities, and more after Class 12th.",
    theme="soft",
    examples=[
        "Best colleges for B.Tech in India",
        "Cutoff rank for CSE in IIT Delhi",
        "Fees and hostel info for VIT Vellore",
        "Which colleges accept CUET score?",
        "Scope of BBA vs B.Com"
    ],
    submit_btn=True
)
```

#Call launch function to execute

```python
iface.launch(share=True)
```
